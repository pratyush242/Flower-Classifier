```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
from sklearn import datasets

```

```python
iris = pd.read_csv("/iris_data.txt",header = None)
iris.columns = ["sepalLengthCm","sepalWidthCm","PetalLengthCm","PetalWidthCm","Species"]
iris_visual = shuffle(iris,random_state=0)
iris_visual.head(10)

```

## visualization of the data set

```python
import seaborn as sns
sns.set()
sns.pairplot(iris[["sepalLengthCm","sepalWidthCm","PetalLengthCm","PetalWidthCm","Species"]],hue = "Species",diag_kind = "kde")
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, y_pred)

sns.heatmap(mat, square = True, annot = True, cbar = False)
plt.xlabel('predicted')
plt.ylabel('true label')
```

Developed a gaussian naive bayes ml model to classify flowers on the basis of distinct sepal and petal features achieveing an overall accuracy of 95\%
